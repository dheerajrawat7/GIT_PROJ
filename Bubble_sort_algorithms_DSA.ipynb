```python
#BUBBLE SORT (ITERETIVE)without swapflag
def bubblesort(lst):
    size=len(lst)

    for indx in range(size):
        for ii in range(0,size-1-indx):
            if lst[ii] > lst[ii+1]:
                lst[ii],lst[ii+1]=lst[ii+1],lst[ii]

            print(f"#pass -{indx},step{ii}:{lst}")

        print('')

def main():
    l1=[10,3,4,25,6,18]
    bubblesort(l1)
    print("final sorted list:")
    for i in range(len(l1)):
        print(l1[i],end=" ")

if __name__=="__main__":
    main()                    
```

```python
#Bubble Sort (iterative)>> complexity O(n^2)
def Bubblesort(lst):
    size = len(lst)

    for indx in range (size):
        swapflag=False
        # traverse through to compare elements
        for ii in range(0,size-1-indx):
            #swap if the e_current > e_next
            if lst[ii] > lst[ii+1]:
                lst[ii],lst[ii+1]=lst[ii+1],lst[ii]
                swapflag=True
            
            #steps
            print(f"pass-{indx},step {ii}:{lst}")
            print(f"swapped:{swapflag}")

        print('')

def main():
    l1=[10, 15, 3, 4, 6, 13, 7]      
    Bubblesort(l1)
    print("final sorted lsit:")
    for i in range(len(l1)):
        print(l1[i], end=' ')

if __name__== '__main__':
    main()                  

```

```python
#BUBBLE SORT (ITERATIVE) 
#BEST CASE Complexity O(n)
def Bubblesort(lst):
    size=len(lst)
    
    for indx in range(size):
        swapflag =False
        for ii in range(0,size-1-indx):
            if lst[ii]>lst[ii+1]:
                lst[ii],lst[ii+1]=lst[ii+1],lst[ii]
                swapflag=True

            #steps
            print(f"#pass-{indx},step{ii}:{lst}",end='')
            print(f"swapped :{swapflag}")

            #remove unnecessary iterations
            if swapflag is False:
                break
            print("")

def main():
    l1=[1,4,6,7,8,10] #this is our sorted list
    Bubblesort(l1)
    print("final sorted list:")
    for i in range(len(l1)):
        print(l1[i],end= ' ')

if __name__=="__main__":
    main()                            
```
